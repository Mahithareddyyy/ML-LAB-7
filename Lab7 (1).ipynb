```python
import numpy as np
import pandas as pd
#Uploading the dataset which is a csv file to python
df=pd.read_excel("embeddingsdata.xlsx")
df
```

```python
#A1
import numpy as np
from sklearn.model_selection import train_test_split
# Selecting any two classes from the dataset
class_A = 2
class_B = 5
#I selected class 2 and 5
X = df.drop(['Label'], axis=1)  #Here we are dropping the target label column and including all the rest of the features
y = df['Label']  #We are taking all the attributes of target label column
# Filtering the data to include only the selected classes i.e. class having label as 2 and class having label as 5
X_selected = X[(y == class_A) | (y == class_B)]
y_selected = y[(y == class_A) | (y == class_B)]
# Splitting the data into a training set and a testing set
X_train, X_test, y_train, y_test = train_test_split(X_selected, y_selected, test_size=0.2, random_state=42)
```

```python
from sklearn import svm
clf = svm.SVC()
#fitting all the features with the svm model created
clf.fit(X_train, y_train)
#Printing the support vectors
clf.support_vectors_
```

```python
#A2
#Calculating and printing the accuracy of SVM
accuracy = clf.score(X_test, y_test)
print("Accuracy of SVM:", accuracy)
```

```python
test_vector = X_test.iloc[0]
# Perform classification for the given feature vector using the trained SVM model
predicted_class = clf.predict([test_vector])
#Printing the predicted class
print(f"The predicted class for the test vector: {predicted_class}")
```

```python
#A3
import numpy as np
import pandas as pd
from sklearn import svm
# Using the predict() function to get predicted class values for the test set
predicted_classes = clf.predict(X_train)
# Calculating accuracy by comparing predicted_classes to the actual y_test labels
correct_predictions = (predicted_classes == y_train)
accuracy = np.mean(correct_predictions)
# Printing the accuracy
print(f"Accuracy: {accuracy:.2f}")
```

```python
#A4
import numpy as np
import pandas as pd
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
kernel_functions = ['linear', 'poly', 'rbf', 'sigmoid']
for kernel in kernel_functions:
    # Creating an SVM classifier with the current kernel function
    clf = svm.SVC(kernel=kernel)
    # Fitting the classifier to the training data
    clf.fit(X_train, y_train)
    # Using the classifier to make predictions on the test set
    y_pred = clf.predict(X_test)
    # Calculating and printing accuracy for the current kernel
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy with '{kernel}' kernel: {accuracy:.2f}")
```
